```python
import yfinance as yf
import warnings 
warnings.filterwarnings('ignore')

# Define the ticker symbol of the stock you want to fetch data for
ticker_symbol = "AAPL"  # Apple Inc. as an example

# Fetch historical data from Yahoo Finance
stock_data = yf.download(ticker_symbol, start="2021-01-01", end="2023-07-25")
```

```python
df = stock_data[["Close"]]
df.head()
```

```python
df['year'] = df.index.year
df['quarter'] = df.index.quarter
df['month'] = df.index.month
df['dayofweek'] = df.index.day_of_week
df['dayofmonth'] = df.index.day
df['dayofyear'] = df.index.day_of_year
df['hour'] = df.index.hour
```

```python
df.head()
```

```python
x_train = df[df.index<"2023-01-01"][df.columns.drop('Close')]
x_test  = df[df.index>="2023-01-01"][df.columns.drop('Close')]
y_train = df[df.index<"2023-01-01"]['Close']
y_test  = df[df.index>="2023-01-01"]['Close']
```

```python
xgb_model = xgboost.XGBRegressor(n_jobs=-1,n_estimators=1000,learning_rate=0.001,booster='gblinear',early_stopping_rounds=10)
```

```python
xgb_model.fit(x_train,y_train,verbose=True,eval_set=[(x_test,y_test)],eval_metric='mae')
```
